Scrape the holidate website's HTML with beautiful soup. Use FOR loop to get 2 years before, this year, 2 years ahead. Convert date to numeric y-m-d. Append to master holiday list. Must be in format Holiday (Date).

```python
from datetime import datetime
import requests
from bs4 import BeautifulSoup
import json

holiday = []

def getHTML(url):
    response = requests.get(url)
    return response.text

for i in range(2020, 2025):
    url = f'https://www.timeanddate.com/holidays/us/{i}'
    print(f'Scraping urls: {url}')

    html = getHTML(f"https://www.timeanddate.com/holidays/us/{i}")
    
    soup = BeautifulSoup(html, 'html.parser')
    
    tablebody = soup.find('table').find('tbody') 

    for j in tablebody.find_all_next('tr'):
        holidate = str(j.find('th', attrs = {'class':'nw'}))
        holidate =(f'{holidate[15:-5]} {i}')
        holiname = str(j.find('a'))
        holiname = holiname[holiname.find('>')+1:-4]
        if holidate != ' 2020' and holiname != '':     #getting blanks in same place in both date and name for some reason. so this takes care of that
            holidate =datetime.strftime(datetime.strptime(holidate,'%b %d %Y'), '%Y-%m-%d') #strp separates, strf re-arranges
            if f'{holiname} ({holidate})' not in holiday: #don't let duplicates in
                holiday.append(f'{holiname} ({holidate})')
            else:
                continue
        else:
            continue

print(holiday)
```
